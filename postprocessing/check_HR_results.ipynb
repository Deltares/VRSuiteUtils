This script reads the results of the HR calculations (design table) from both 2023 and 2100 for a specific cross section and makes a plot of water levels versus beta

```python
# import the necessary packages
import matplotlib.pyplot as plt
import os
from preprocessing.step4_build_sqlite_db.read_intermediate_outputs import read_design_table
```

```python
# define the path to the folder that contains the 2023 and 2100 folder
hr_results_path = r"c:\VRM\20240618_bekleding_renske\intermediate_results\HR_results\waterlevel"

years = [2023, 2100]
year_path_current = hr_results_path + "\\" + str(years[0])
year_path_future = hr_results_path + "\\" + str(years[1])
print(year_path_current)
print(year_path_future)

# create a list of all folder names in year_path_current

folder_names_current = os.listdir(year_path_current)
folder_names_future = os.listdir(year_path_future)
# check if all folders in folder_names_current are also present in year_path_future, if so, return True
# if not, return False
if all(x in folder_names_future for x in folder_names_current):
    print("All folders are present in both years")
    print("Cross section folders: ", folder_names_current)
else:
    print("Not all folders are present in both years")
    

```

```python
# loop through the folder names and read the .txt file starting with "DESIGNTABLE" in both 2023 and 2100, using read_design_table function
for folder in folder_names_current:
    print("folder: ", folder)
    # find the file that starts with "DESIGNTABLE" in the current folder
    design_table_filename = [f for f in os.listdir(year_path_current + "\\" + folder) if f.startswith("DESIGNTABLE")][0]
    print("Design table filename: ", design_table_filename)
    design_table_current = read_design_table(year_path_current + "\\" + folder + "\\" + design_table_filename)
    design_table_future = read_design_table(year_path_future + "\\" + folder + "\\" + design_table_filename)
    print("Design table current: ", design_table_current)
    print("Design table future: ", design_table_future)
    # plot the design table
    plt.plot(design_table_current["Value"], design_table_current["Beta"], label="2023")
    plt.plot(design_table_future["Value"], design_table_future["Beta"], label="2100")
    plt.xlabel("water level [m+NAP]")
    plt.ylabel("Beta")
    plt.grid()
    plt.legend()
    # save plot to year_path_current + "\\" + folder + "\\" + "design_table.png"
    plt.savefig(year_path_current + "\\" + folder + "\\" + "design_table.png")
    plt.show()
```
